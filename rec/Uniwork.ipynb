# read MyCsv file ‚¨áÔ∏è


```python
#read MyCsv file
df = pd.read_csv('uniList')
```

# combine_columns F ‚¨áÔ∏è


```python
#combine_columns F

def combine_columns(row):
    return row['University'] + row['Country']
```

# merge ‚¨áÔ∏è

```python

df['C'] = df.apply(combine_columns, axis=1)
```

### Scrap Row Data Using OpenAi ToolS üÜïüóûÔ∏è
#### Scrap the majors using Ai ‚¨áÔ∏è

```python
#Scrap Data Using OpenAi ToolS üÜïüóûÔ∏è

# Scrap the majors using Ai

import openai
import time
openai.api_key = "key"


unidata={"scrap":["one",2,3.5]}
saved=[]
n=0
for i in df['C']:

    response = openai.ChatCompletion.create(
      model="gpt-3.5-turbo",
      messages=[
        {
          "role": "system",
          "content": "You will help me more then u can"
        },
        {
          "role": "user",
          "content": f"Without any introductions or details or ant type of notes, provided me with 10 lists of technical major available at this university :{i}"
        }
      ],

      temperature=1,
      max_tokens=256,
      top_p=1,
      frequency_penalty=0,
      presence_penalty=0
    ) 
    ###################################
    saved.append(response.choices[0]["message"]["content"])
    unidata[f"{i}"]= response.choices[0]["message"]["content"]
    ###################################
    print(response.choices[0]["message"]["content"])
    
    n+=1
    print(n)
    
    
    
    
```
